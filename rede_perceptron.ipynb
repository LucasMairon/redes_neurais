```python
import numpy as np
import matplotlib.pyplot as plt


class Perceptron:
    # Construtor da classe
    def __init__(self, n_inputs, taxa_aprendizado=0.1):
      self.pesos = [0] * n_inputs
      self.bias = 0
      self.eta = taxa_aprendizado

    # Função degrau
    def ativacao(self, entrada):
      if entrada >= 0:
        return 1
      else:
        return 0

    # Função de previsão
    def prever(self, x):
      """
      Calcula a saida da rede
      """
      soma = 0
      for i in range(len(x)):
        soma += x[i] * self.pesos[i]

      soma += self.bias
      y = self.ativacao(soma)
      return y

    def treinar(self, entrada, saida_desejada, epocas):
      for epoca in range(epocas):
        for count in range(len(entrada)):

          saida_rede = self.prever(entrada[count])
          erro = saida_desejada[count] - saida_rede

          for i in range(len(self.pesos)):
            self.pesos[i] += self.eta * erro * entrada[count][i]
          self.bias += self.eta * erro

        print(f"Epocas: {epoca}, Pesos: {self.pesos}")
```

```python
entradas = [[0, 0],[0, 1],[1, 0],[1, 1]]
saida_desejada = [0, 1, 1, 1]


rede_percepton = Perceptron(n_inputs=2, taxa_aprendizado=0.1)
rede_percepton.treinar(entradas, saida_desejada, 20)

for entrada in entradas:
  print(f' Entrada: {entrada} saida prevista: {rede_percepton.prever(entrada)}')
```
