```python
import numpy as np


def ativacao(entrada_funcao):
  """
  Função de ativação degrau para o neuronio
  """
  if entrada_funcao >= 0:
    return 1
  else:
    return 0

def neuronio(entrada, peso, bias):
  """
    Função do neuronio:
    entrada: x
    peso: w
    bias: b

  """
  # chamar a função dot da numpy para calcular o produto escalar
  u = np.dot(entrada, peso) + bias
  y = ativacao(u)
  return y
```

```python
entrada = np.array([1, 0])
pesos = np.array([0.6, 0.2])
bias = -0.5

saida = neuronio(entrada, pesos, bias)
print(f'Saida do neuronio:{saida}')
```
